```python
import os
import requests

# Mapeo de meses en catal√°n
i2m = list(zip(range(1, 13), ['Gener', 'Febrer', 'Marc', 'Abril', 'Maig', 'Juny', 
                              'Juliol', 'Agost', 'Setembre', 'Octubre', 'Novembre', 'Desembre']))

# Crear la carpeta de descargas si no existe
download_folder = "data"
os.makedirs(download_folder, exist_ok=True)

# Rango de a√±os (2024 a 2020)
for year in range(2024, 2019, -1):
    for month, month_name in i2m:
        # Si el mes es mayor a 5 en el a√±o 2024, saltarlo
        if month > 5 and year > 2023:
            continue

        # Nombre del archivo de destino
        filename = f"{year}{month:02d}{month_name}_BicingNou_ESTACIONS.7z"
        filepath = os.path.join(download_folder, filename)

        # URL de descarga
        url = f"https://opendata-ajuntament.barcelona.cat/resources/bcn/BicingBCN/{year}_{month:02d}_{month_name}_BicingNou_ESTACIONS.7z"

        # Verificar si la URL existe antes de descargar
        response = requests.head(url)
        if response.status_code == 200:
            print(f"Descargando: {filename}")
            file_response = requests.get(url, stream=True)
            with open(filepath, "wb") as file:
                for chunk in file_response.iter_content(chunk_size=8192):
                    file.write(chunk)
            print(f"‚úÖ Descargado correctamente: {filename}")
        else:
            print(f"‚ùå No se encontr√≥: {url} (C√≥digo {response.status_code})")

print("Proceso de descarga finalizado.")
```

```python
import os
import subprocess

# Asegurarse de que la carpeta 'data' exista
os.makedirs("data", exist_ok=True)

# Lista de tuplas que asocian n√∫mero de mes y su nombre en catal√°n
i2m = list(zip(
    range(1, 13),
    ['Gener', 'Febrer', 'Marc', 'Abril', 'Maig', 'Juny', 
     'Juliol', 'Agost', 'Setembre', 'Octubre', 'Novembre', 'Desembre']
))

for year in range(2024, 2019, -1):
    for month, month_name in i2m:
        # Para 2024 (o a√±os mayores a 2023) se descargan solo los meses hasta mayo
        if (month > 5) and (year > 2023):
            continue

        # Construir el nombre del archivo y la URL
        file_name = f"data/{year}_{month:02d}_{month_name}_BicingNou_ESTACIONS.7z"
        file_path = os.path.join("data", file_name)
        url = f"https://opendata-ajuntament.barcelona.cat/resources/bcn/BicingBCN/{file_name}"
        
        # Descargar el archivo .7z usando curl
        os.system(f'curl -L -o "{file_path}" "{url}"')
        
        # Extraer el contenido del archivo .7z usando 7z
        # Se asume que 7z est√° en el PATH; si no, se puede usar la ruta completa (ver ejemplo abajo)
        try:
            subprocess.run(["7z", "x", file_path, "-odata"], check=True)
        except FileNotFoundError:
            # Si no se encuentra "7z" en el PATH, especifica la ruta completa:
            subprocess.run(["C:\\Program Files\\7-Zip\\7z.exe", "x", file_path, "-odata"], check=True)
```
