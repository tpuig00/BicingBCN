```python
#DESCARGA UTILITZADA PER CLARA (sino no em funcionava)
import os
import requests
import py7zr  # Librería para extraer archivos .7z

# Crear la carpeta data si no existe
output_folder = "data"
os.makedirs(output_folder, exist_ok=True)

i2m = list(zip(range(1, 13), ['Gener', 'Febrer', 'Marc', 'Abril', 'Maig', 'Juny', 'Juliol', 
                              'Agost', 'Setembre', 'Octubre', 'Novembre', 'Desembre']))

for year in range(2024, 2019, -1):  
    for month, month_name in i2m:
        filename = f"{year}_{month:02d}_{month_name}_BicingNou_ESTACIONS.7z"
        file_path = os.path.join(output_folder, filename)
        url = f"https://opendata-ajuntament.barcelona.cat/resources/bcn/BicingBCN/{filename}"

        print(f"📥 Descargando: {filename}...")

        # Descargar el archivo
        try:
            response = requests.get(url, stream=True, timeout=10)
            response.raise_for_status()  # Lanza un error si la respuesta no es 200

            with open(file_path, 'wb') as f:
                for chunk in response.iter_content(1024):
                    f.write(chunk)

            print(f"✅ Descarga completada: {filename}")

            # Extraer el archivo .7z
            try:
                with py7zr.SevenZipFile(file_path, mode='r') as archive:
                    archive.extractall(path=output_folder)
                print(f"📂 Archivo extraído en {output_folder}")

                # Eliminar el archivo .7z después de extraer
                os.remove(file_path)

            except Exception as e:
                print(f"⚠️ Error al extraer {filename}: {e}")

        except requests.exceptions.RequestException as e:
            print(f"❌ Error al descargar {filename}: {e}")
```
